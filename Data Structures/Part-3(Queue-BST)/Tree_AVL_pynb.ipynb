```python
class TreeNode:
  def __init__(self,data, children=None):
    self.data = data
    self.children = children if children is not None else []
  def __str__(self,level=0):
    result = ' '* level + str(self.data) + '\n'
    for child in self.children:
      result += child.__str__(level+1)
    return result
  def add_child(self, node):
    self.children.append(node)
main_node = TreeNode("drinks")
cold = TreeNode("cold")
hot = TreeNode("Hot")
sprite = TreeNode("sprite")
fanta = TreeNode("fanta")
cofee = TreeNode("cofee")
tea = TreeNode("tea")
main_node.add_child(cold)
main_node.add_child(hot)
cold.add_child(sprite)
cold.add_child(fanta)
hot.add_child(cofee)
hot.add_child(tea)
print(main_node)
```
